```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


class SimpleANNRegressor:
    def __init__(self):
        self.model = Sequential([
            Dense(2, activation='relu', input_shape=(2,)),  # hidden layer (2 neurons)
            Dense(1, activation='linear')                   # output layer
        ])

        self.model.compile(
            optimizer=Adam(learning_rate=0.01),
            loss='mse'
        )

    def train(self, X, y, epochs=200):
        self.model.fit(X, y, epochs=epochs, verbose=1)

    def predict(self, x1, x2):
        X = np.array([[x1, x2]])
        return self.model.predict(X, verbose=0)[0][0]


# ---------------------------
# Load your data
# ---------------------------

df = pd.read_csv("data.csv")

X = df.iloc[:, 0:2].values   # two input columns
y = df.iloc[:, 2].values    # target column


# ---------------------------
# Train model
# ---------------------------

ann = SimpleANNRegressor()
ann.train(X, y)


# ---------------------------
# Take input and predict
# ---------------------------

x1 = float(input("Enter x1: "))
x2 = float(input("Enter x2: "))

result = ann.predict(x1, x2)

print("Predicted value =", result)
```
